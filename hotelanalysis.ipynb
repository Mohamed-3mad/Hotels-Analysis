## import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

## importing & Inspecting data

```python
hotels = pd.read_csv('hotels.csv')
hotels.head()
```

## Cleaning data

```python
hotels.drop(['agent','company'], axis = 1,inplace = True)
hotels.dropna(inplace = True)
```

```python
hotels.shape
```

```python
hotels.describe()
```

```python
hotels['adr'].plot.box()
```

```python
hotels = hotels[hotels['adr']<5000]
```

```python
hotels.describe()
```

## Analyzing data

```python
cancelled_perc = hotels['is_canceled'].value_counts(normalize = True)
cancelled_perc
```

```python
plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not Canceled','Canceled'], hotels['is_canceled'].value_counts(), edgecolor='k')
```

```python
plt.figure(figsize=(5,6))
sns.countplot(x='hotel', hue='is_canceled', data = hotels, palette = 'Pastel1')
plt.title('Cancellation rates in City hotel & Resort hotel',fontweight='bold', size=20)
plt.show()
```

## The City Hotel has the highest booking rate comparted to resort hotel
## The Resort Hotel has the lowest cancellation rate comparted to city hotel.

```python
city_hotel = hotels[hotels['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = hotels[hotels['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel= resort_hotel.groupby('reservation_status_date')[['adr']].mean()
resort_hotel
```

```python
city_hotel= city_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel
```

```python
plt.figure(figsize= (20,8))
plt.title('Avrage daily rate in city & resort hotel', fontsize= 30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label = 'Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
hotels['month'] = hotels['reservation_status_date'].dt.month
plt.figure(figsize= (15,7))
sns.countplot(x='month',hue='is_canceled', data=hotels,palette='Pastel1')
plt.title('Reservation status per month', fontweight = 'bold', size=20)
plt.xlabel('month',fontweight ='bold',size = 15)
plt.ylabel('no. of reservations',fontweight ='bold',size = 15)
plt.legend(['not_canceled', 'canceled'])
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.title('ADR per Month', fontsize = 30)
data = hotels[hotels['is_canceled']==1].groupby('month')['adr'].sum().reset_index()
sns.barplot(x='month' , y='adr',data=data)
plt.show()
```

```python
cancelled_data = hotels[hotels['is_canceled']==1]
top_10_countries = canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with the most cancellations',color ='black')
plt.pie(top_10_countries, autopct='%.2f',labels = top_10_countries.index)
plt.show()
```

```python
hotels['market_segment'].value_counts(normalize = True) 
```

```python
cancelled_data['market_segment'].value_counts(normalize = True) 
```

```python
cancelled_resv_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_resv_adr.reset_index(inplace = True)
cancelled_resv_adr.sort_values('reservation_status_date',inplace = True)

not_cancelled_data = hotels[hotels['is_canceled']==0]
not_cancelled_resv_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_resv_adr.reset_index(inplace = True)
not_cancelled_resv_adr.sort_values('reservation_status_date',inplace = True)

plt.figure(figsize= (20,6))
plt.title('Avrage daily rate', fontsize= 40,color = 'Black')
plt.plot(not_cancelled_resv_adr['reservation_status_date'],not_cancelled_resv_adr['adr'],label = 'not_cancelled')
plt.plot(cancelled_resv_adr['reservation_status_date'],cancelled_resv_adr['adr'],label = 'cancelled')
plt.legend(fontsize = 20)
plt.show()
```
